```python
def get_sentiment(final_sentence1):
    s1 = flair.data.Sentence(final_sentence1)
    flair_sentiment.predict(s1)
    total_sentiment_s1 = s1.labels
    if str(total_sentiment_s1[0])[0:8] == "POSITIVE":
        sent1 = float(str(total_sentiment_s1[0])[10:len(str(total_sentiment_s1[0]))-2])
    else:
        sent1 = -1 * float(str(total_sentiment_s1[0])[10:len(str(total_sentiment_s1[0]))-2])
    return sent1
```

```python
def trial(df_test):
    stop_words = set(stopwords.words('english')) 
    ps = PorterStemmer()
    flair_sentiment = flair.models.TextClassifier.load('en-sentiment')
    id_list = []
    results_list = []

    for i in tqdm(range(df_test.shape[0])):
        sentence1 = word_tokenize(df_test.at[i,'InputSentence1'])
        sentence2 = word_tokenize(df_test.at[i,'InputSentence2'])
        sentence3 = word_tokenize(df_test.at[i,'InputSentence3'])
        sentence4 = word_tokenize(df_test.at[i,'InputSentence4'])
        random1 = word_tokenize(df_test.at[i,'RandomFifthSentenceQuiz1'])
        random2 = word_tokenize(df_test.at[i,'RandomFifthSentenceQuiz2'])

        filtered_sentence1 = [ps.stem(w) for w in sentence1 if not w in stop_words] 
        filtered_sentence2 = [ps.stem(w) for w in sentence2 if not w in stop_words] 
        filtered_sentence3 = [ps.stem(w) for w in sentence3 if not w in stop_words] 
        filtered_sentence4 = [ps.stem(w) for w in sentence4 if not w in stop_words] 
        filtered_random1 = [ps.stem(w) for w in random1 if not w in stop_words] 
        filtered_random2 = [ps.stem(w) for w in random2 if not w in stop_words] 


        final_sentence1 = ' '.join(filtered_sentence1)
        final_sentence2 = ' '.join(filtered_sentence2)
        final_sentence3 = ' '.join(filtered_sentence3)
        final_sentence4 = ' '.join(filtered_sentence4)
        final_random1 = ' '.join(filtered_random1)
        final_random2 = ' '.join(filtered_random2)


        sent1 = get_sentiment(final_sentence1)
        sent2 = get_sentiment(final_sentence2)
        sent3 = get_sentiment(final_sentence3)
        sent4 = get_sentiment(final_sentence4)
        sentr1 = get_sentiment(final_random1)
        sentr2 = get_sentiment(final_random2)
        
        print("{sd}")

        total_sent = sent1 * 0.25 + sent2 * 0.5 + sent3 * 0.75 + sent4 * 1 
        #total_sent = sent1 + sent2 + sent3 + sent4
        
        id_list.append(df_test.at[i,'InputStoryid'])
        
        if (total_sent >= 0) and (sentr1 >= 0):
            results_list.append(1)
        elif (total_sent >= 0) and (sentr2 >= 0):
            results_list.append(2)
        elif (total_sent < 0) and (sentr1 < 0):
            results_list.append(1)
        else:
            results_list.append(2)
            
    return id_list, results_list


    
```

```python
id_list, results_list = trial(df_dev)

count = 0
for i in tqdm(range(df_dev.shape[0])):
    if df_dev.at[i,'AnswerRightEnding'] == results_list[i]:
        count = count + 1

print("Accuracy: " + str(float(count/df_dev.shape[0])))
```
